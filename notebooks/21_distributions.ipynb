# Distributions

This notebook presents different distributions for discrete choice modelling

### 1. Uniform

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_dist(dist, title=None):
    plt.plot(dist)
    plt.title(title)
    plt.show()

def show_stats(dist):
    print(f"Mean: {dist.mean()}, Var: {dist.var()}")
```

```python
# Sample N numbers from a uniform distribution U(0,1)
N = 1000
d_uniform = np.random.uniform(size=N)
show_dist(d_uniform, "Uniform(0,1)")
```

```python
# Sample N numbers from a uniform distribution U(a,b)
# U(a,b) = (b-a)*r + a with r is from U(0,1)
N = 1000
a,b = 4,8
r = np.random.uniform(size=N)
ab_uniform = (b-a)*r + a 
show_stats(ab_uniform)
show_dist(ab_uniform, f"Uniform({a},{b})")
```

### 2. Normal distribution

```python
# IF r1 and r2 are independent draws from U(0,1); 
# THEN s1 = sqrt(-2(ln(r1)))*sin(2*pi*r2), and s2 = sqrt(-2ln(r1))*cos(2*pi*r2) are
# independent draws from N(0, 1)
N = 1000
r1 = np.random.uniform(size=N)
r2 = np.random.uniform(size=N)
s1 = np.sqrt(-2*np.log(r1)) * np.sin(2*math.pi*r2)
s2 = np.sqrt(-2*np.log(r1)) * np.cos(2*math.pi*r2)
s1 is s2
```

```python
show_stats(s1)
show_dist(s1,"N(0,1)")
```

```python
show_stats(s2)
show_dist(s2,"N(0,1)")
```

```python
# IF r is a draw from N(0,1) THEN s = br + a is a draw from N(a,b^2)
N = 1000
a,b = 5,3
r = np.random.normal(size=N)
s = a + b*r
show_stats(s)
show_dist(s)
```

### 3. Log Normal distribution

```python
# If r is a draw from N(a,b^2), then e^r is a draw from a log_normal LN(a,b^2)
# new mean: e^(a+b^2/2), var: e^(2a + b^2) * (e^(b^2) - 1)
r = s
exp_r = np.exp(r)
show_stats(exp_r)
show_dist(exp_r)
```
